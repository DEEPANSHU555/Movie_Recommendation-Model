```python
#STEP 1 IMPORT LIBRARIES
import pandas as pd
import numpy as np
```

```python
#STEP 2 IMPORT DATASET
df= pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
df.fillna(value= "none",inplace = True)
df.head(239)
```

```python
#STEP 3 FEATURE SELECTION
df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']]
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['Movie_Genre'] + ' ' +df_features['Movie_Keywords'] + ' ' +df_features['Movie_Tagline'] + ' ' +df_features['Movie_Cast'] + ' ' +df_features['Movie_Director']
```

```python
x
```

```python
x.shape
```

```python
#STEP 4  TEXT CONVERSION TO TOKENS
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
# STEP 5 SIMILARITY SCORE USING COSINE
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

```python
#STEP 6 GET MOVIE NAME FROM USER AND VALIDATE CLOSEST SPELLING
Favourite_Movie_Name = input("Enter Your Favourite Movie Name:")
```

```python
All_Movie_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
Find_Close_Match =difflib.get_close_matches(Favourite_Movie_Name,All_Movie_Title_List)
Close_Match =Find_Close_Match[0]
Index_Of_Movie =df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similarity_Score[Index_Of_Movie]))
Sorted_Similar_Movie = sorted(Recommendation_Score,key= lambda x:x[1],reverse = True)
print('Top 10 Movies Suggested for you :')
i=1
for movie in Sorted_Similar_Movie:
  index=movie[0]
  Title_From_Index = df[df.Movie_ID ==index]['Movie_Title'].values
  if(i<11):
    print(i,'.', Title_From_Index)
    i=i+1
```
